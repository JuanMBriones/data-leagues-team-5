Se guardan las url's de los archivos solicitados en las variables `u1011`, `u1112` y `u1213` y posteriormente se prodece a descargar los archivos solicitados con el comando `download.file()`

```r
u1011 <- "https://www.football-data.co.uk/mmz4281/1920/SP1.csv"
u1112 <- "https://www.football-data.co.uk/mmz4281/1819/SP1.csv"
u1213 <- "https://www.football-data.co.uk/mmz4281/1718/SP1.csv"

download.file(url = u1011, destfile = "esp20.csv", mode = "wb")
download.file(url = u1112, destfile = "esp19.csv", mode = "wb")
download.file(url = u1213, destfile = "esp18.csv", mode = "wb")
```

Se importan los dataframes descargados como elementos de una lista llamada `lista`. Esto se puede hacer para todos los dataframes en una sentada con el comando `lapply()`.

```r
(lista <- lapply(dir(), read.csv))
```

Se visualizan los datos y la estructura del los dataframes 

```r
head(lista[[1]]); head(lista[[2]]); head(lista[[3]])
```

```r
str(lista[[1]]); str(lista[[2]]); str(lista[[3]])
```

```r
View(lista[[1]]); View(lista[[2]]); View(lista[[3]])
```

```r
summary(lista[[1]]); summary(lista[[2]]); summary(lista[[3]])
```

Se extraen los campos Date, HomeTeam, AwayTeam, FTHG, FTAG y FTR de cada uno de los dataframes, nuevamente con el comando `lapply()` y `select()`. Además, se cambia el formato del campo Date por el de fecha.

```r
nueva_lista <- lapply(lista, select, c("Date", "HomeTeam", "AwayTeam", "FTHG", "FTAG", "FTR"))
nueva_lista <- lapply(nueva_lista, mutate, Date = as.Date(Date, "%d/%m/%Y"))
```

Por último, con la ayuda de la función `do.call()` y `rbind()` unimos los datos de los 3 dataframes en uno sólo. Se exporta este `data_partidos_18_21` con la función `write.csv()`.

```r
data_partidos_18_21 <- do.call(rbind, nueva_lista)
write.csv(data_partidos_18_21, "Partidos_18_21.csv", row.names = FALSE)
tail(data_partidos_18_21)
```
